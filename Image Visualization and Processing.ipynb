This Python script showcases image visualization and basic processing techniques using the popular computer vision libraries, OpenCV and Matplotlib. The script covers functionalities such as displaying images with Matplotlib and OpenCV, plotting image histograms, visualizing RGB channels, and resizing images.

```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
# This function is designed to display an image using the Matplotlib library.
# fig, ax  creates a Matplotlib figure (fig) and axis (ax). The figsize parameter controls the size of the figure.
# If axis_off is set to True, this line turns off the axis labels, providing a clean image display without axis ticks.
def display_image_with_matplotlib(image, title=None, figsize=(8, 8), cmap=None, axis_off=True):
    fig, ax = plt.subplots(figsize=figsize)
    ax.imshow(image, cmap=cmap)
    if axis_off:
        ax.axis('off')
    if title:
        ax.set_title(title)
    plt.show()
```

```python
# This function is designed to display an image using the OpenCV library
# the if statement checks if the input image has three channels (indicating an RGB image). If true, it converts the image from RGB to BGR format using cv2.cvtColor. OpenCV uses BGR as its default color order.
# waitKey(0) tells that This function waits for a key event. The argument 0 indicates that the window will wait indefinitely until a key is pressed.
def display_image_with_cv2(image,title):
    if len(image.shape) == 3 and image.shape[2] == 3:
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    cv2.imshow(title, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
# This function is designed to plot the histogram of pixel intensities in an image using the Pandas library and Matplotlib. 
# pd.Series(image.flatten()): This line flattens the 2D or 3D image array into a 1D array, making it suitable for histogram plotting ans converts the 1D NumPy array into a Pandas Series, which is a one-dimensional labeled array.
#  The kind='hist' parameter specifies the type of plot.
# he function provides a convenient way to visualize the distribution of pixel intensities in an image using a histogram. It can be helpful for understanding the overall brightness and contrast characteristics of the image.
def plot_image_histogram(image):
    pd.Series(image.flatten()).plot(kind='hist')
    plt.show()
```

```python
# This function is designed to display the individual RGB channels of an image using Matplotlib
# plt.subplots(1, 3, figsize=(15, 5)): This line creates a Matplotlib figure with three subplots arranged horizontally. Each subplot corresponds to one of the RGB channels.
# The loop iterates over the three RGB color channels: Red, Green, and Blue.
def display_rgb_channels(image, title=None):
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    for i, color in enumerate(['Reds', 'Greens', 'Blues']):
        axs[i].imshow(image[:,:,i], cmap=color)
        axs[i].set_title(f'{color[:-1]} channel')
    plt.show()
```

```python
# This function is designed to compare and display images using both OpenCV and Matplotlib.
# axs[0] displays the image using OpenCV in the first subplot.
# axs[1] displays the image using Matplotlib in the second subplot.
def display_images_cv2_vs_matplotlib(cv2_image, matplotlib_image):
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))
    axs[0].imshow(cv2_image)
    axs[1].imshow(matplotlib_image)
    axs[0].set_title('CV2 Image')
    axs[1].set_title('Matplotlib Image')
    plt.show()
```

```python
# This function is designed to resize an image using OpenCV and then display the resized image using Matplotlib
# condition checks if the target_size parameter is specified. If so, it resizes the image
def resize_and_display_image(image, target_size=None, interpolation=cv2.INTER_LINEAR):
    if target_size:
        resized_image = cv2.resize(image, target_size, interpolation=interpolation)
    else:
        resized_image = image
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.imshow(resized_image)
    ax.axis('off')
    plt.show()
```

```python
# load imag both with matplotlib and cv2
image_plt= plt.imread('party.jpg')
image_cv2= cv2.imread('party.jpg')
image_cv2=cv2.cvtColor(image_cv2, cv2.COLOR_BGR2RGB)
```

```python
display_image_with_matplotlib(image_plt, title='Matplotlib Image')
```

```python
display_image_with_cv2(image_cv2,title='CV2 Image')
```

```python
plot_image_histogram(image_cv2)
```

```python
display_rgb_channels(image_plt, title='RGB Channels of Cat Image')
```

```python
display_images_cv2_vs_matplotlib(image_cv2, image_plt)
```

```python
resize_and_display_image(image_cv2, target_size=(200, 200))
```

```python

resize_and_display_image(image_cv2, target_size=(1500, 2000), interpolation=cv2.INTER_CUBIC)

```
